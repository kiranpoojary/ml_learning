```python
import pandas as pd 
```

```python
df=pd.read_csv("town_area.csv")
df
```

# Using  Dummy variable

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
#concat df and dummies
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
#remove town and one of dummies col to avoid dummy variable trap
#but in linear regression the dummy variable drop itself but its good practice to drop ourself
final=merged.drop(['town'],axis='columns')
final
```

```python
#x and y for training
final=final.drop(['west windsor'],axis='columns')
final
```

```python
X = final.drop('price', axis='columns')
X
```

```python
y=final.price
y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
#accuracy
model.score(X,y)
```

```python
model.predict([[3400,0,0]])
```

# Using One Hot Encoder

```python

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.town=le.fit_transform(dfle.town)
dfle
```

```python
X=dfle[['town','area']].values
X
```

```python
y=dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('town',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
X=ct.fit_transform(X)
X
```

```python
#drop
X=X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]])
```

```python
model.score(X,y)
```
